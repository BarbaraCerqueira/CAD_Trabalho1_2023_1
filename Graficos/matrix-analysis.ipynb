### Analise da multiplicação matriz-vetor para diferentes ordens de loops em C e Fortran

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.preprocessing import minmax_scale
```

---

#### Multiplicação matriz-vetor em C

```python
# Tamanhos N de sistema testados
x = [64,512,1024,4096,8192,16384,20000,22000,24000,26000]

# Tempos de CPU para matrix-multiplier.c
y1 = [0.000005,
    0.000462,
    0.001555,
    0.027021,
    0.100912,
    0.390086,
    0.591828,
    0.750809,
    0.875220,
    1.012646]

# Tempos de CPU para matrix-multiplier-v2.c
y2 = [0.000004,
    0.000558,
    0.002365,
    0.104276,
    0.540198,
    3.850278,
    6.792066,
    10.959658,
    12.350653,
    14.953834]

# Plotando os pontos e a curva da função n^2 no mesmo gráfico
plt.plot(x, y1, 'o', label='Linha a linha')
plt.plot(x, y2, 'o', label='Coluna a coluna')
plt.xlabel('Tamanho do Sistema (N)')
plt.ylabel('Tempo de CPU (s)')
plt.legend()
plt.title('Multiplicação Matriz-vetor em C')
plt.show()
```

```python
# Curva da função n^2 com ordenada reescalonada
n = np.arange(0, 26000, 1)
t = n ** 2
t_scaled1 = minmax_scale(t, feature_range = (np.min(y1),np.max(y1)))
t_scaled2 = minmax_scale(t, feature_range = (np.min(y2),np.max(y2)))

# Plotando os pontos e a curva da função n^2 nos mesmos gráficos
figure, axis = plt.subplots(1, 2, figsize = (14,6))

axis[0].plot(x, y1, 'o', label='Linha a linha')
axis[0].plot(n, t_scaled1, '-', label='Curva O(N^2) ajustada')
axis[0].set_xlabel('Tamanho do Sistema (N)')
axis[0].set_ylabel('Tempo de CPU (s)')
axis[0].set_title("Comparação entre O(N^2) e os dados de matrix-multiplier.c")
axis[0].legend()

axis[1].plot(x, y2, 'o', label='Coluna a coluna')
axis[1].plot(n, t_scaled2, '-', label='Curva O(N^2) ajustada')
axis[1].set_xlabel('Tamanho do Sistema (N)')
axis[1].set_ylabel('Tempo de CPU (s)')
axis[1].set_title("Comparação entre O(N^2) e os dados de matrix-multiplier-v2.c")
axis[1].legend()

plt.show()
```

#### Multiplicação matriz-vetor em Fortran

```python
# Tamanhos N de sistema testados
x = [64,512,1024,4096,8192,16384,20000,22000,24000]

# Tempos de CPU para matrix-multiplier.f90
y3 = [0.0000050,
    0.0006300,
    0.0065100,
    0.1342340,
    0.5330320,
    2.9421639,
    5.9398270,
    7.5960732,
    9.1372833]

# Tempos de CPU para matrix-multiplier-v2.f90
y4 = [0.0000030,
    0.0002300,
    0.0006850,
    0.0106748,
    0.0416300,
    0.1666219,
    0.2530491,
    0.3218098,
    0.3839512]

# Plotando os pontos e a curva da função n^2 no mesmo gráfico
plt.plot(x, y3, 'o', label='Linha a linha')
plt.plot(x, y4, 'o', label='Coluna a coluna')
plt.xlabel('Tamanho do Sistema (N)')
plt.ylabel('Tempo de CPU (s)')
plt.legend()
plt.title('Multiplicação Matriz-vetor em Fortran')
plt.show()
```
