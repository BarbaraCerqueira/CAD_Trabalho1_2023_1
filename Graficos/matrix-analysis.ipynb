```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.preprocessing import minmax_scale
```

#### Multiplicação linha por linha em C (matrix-multiplier.c)

```python
# Pontos a serem plotados
x = [64,512,1024,4096,8192,16384,20000,22000,24000,26000]

y = [0.000005,
    0.000462,
    0.001555,
    0.027021,
    0.100912,
    0.390086,
    0.591828,
    0.750809,
    0.875220,
    1.012646]

y2 = [0.000004,
    0.000558,
    0.002365,
    0.104276,
    0.540198,
    3.850278,
    6.792066,
    10.959658,
    12.350653,
    14.953834]

x = x / np.linalg.norm(x) 
y = y
y2 = y2

# Curva da função n^2
n = np.arange(0, 1, 0.1)
t = n ** 2

# Plotando os pontos e a curva da função n^2 no mesmo gráfico
plt.plot(x, y, 'o', label='Amostras')
plt.plot(x, y2, 'o', label='Amostras')
plt.plot(n, t, '-', label='Curva O(N^2)')
plt.xlabel('N (Tamanho do Sistema)')
plt.ylabel('Tempo de Execução (s)')
plt.legend()
plt.show()
```

#### Multiplicação coluna por coluna em C (matrix-multiplier-v2.c)
